```python
import matplotlib.pyplot as plt

# Bar chart
x = ['A', 'B', 'C']
y = [10, 5, 8]
colors = ['skublue', 'orange', 'green']
plt.bar(x, y, color=colors)
plt.title("Bar Chart")
plt.show()

# Scatter plot
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.show()

# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4]
plt.hist(data, bins=4, color='purple')
plt.title("Histogram")
plt.show()

# Pie chart
labels = ['Apples', 'Banana', 'Cherries']
sizes = [30, 45, 25]
colors = ['red', 'yellow', 'pink']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title("Pie Chart")
plt.show()
```
